2. Buatlah sebuah fungsi untuk membaca file json yang disediakan dan mengambil data
"cases" covid di indonesia. return menjadi satu buah list

```python
import json

def read_covid():
    try:
        with open("/content/drive/MyDrive/PKS DIgital School/2.json", 'r') as file:
            data = json.load(file)
            hasil = [i["Cases"] for i in data]
            return hasil
    except FileNotFoundError:
        print("File nggak ketemukan")
    except IOError:
        print("Error pembaca file")
    except KeyError:
        print("Data tidak JSON tidak ditemukan")


cases = read_covid()
print(cases)


```
