# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()



```python
#code here
df = df.dropna()

# Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()
grouped_data = df.groupby('species').mean()

print(grouped_data)
```

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
*   Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g
*   Jelaskan insight yang diperoleh dari ke dua grafik tersebut.





```python
#code here
# Pie plot dari dataset hasil grouping untuk kolom 'flipper_length_mm'
plt.figure(figsize=(8, 6))
grouped_data['flipper_length_mm'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Persentase Flipper Length by Species")
plt.ylabel('')
plt.show()

# Histogram plot dari dataset awal untuk kolom 'body_mass_g'
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='body_mass_g', bins=10, kde=True)
plt.title("Histogram Body Mass")
plt.xlabel("Body Mass (g)")
plt.ylabel("Count")
plt.show()
```

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
#code here
sns.catplot(data=df, x='sex', y='bill_length_mm', kind='bar')
plt.title('Catplot: Bill Length by Sex')
plt.show()

# Grafik Scatter Plot
sns.scatterplot(data=df, x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm')
plt.title('Scatter Plot: Body Mass vs Flipper Length (Colored by Bill Length)')
plt.show()
```

1. Grafik Catplot menunjukkan perbedaan rata-rata panjang paruh (bill length) antara jenis kelamin jantan dan betina pada penguin. Dalam konteks ini, kita dapat melihat apakah terdapat pola atau korelasi antara panjang paruh dengan massa tubuh dan panjang sirip penguin. Misalnya, jika terdapat pola di mana penguin dengan panjang paruh tertentu memiliki massa tubuh yang lebih besar atau panjang sirip yang lebih panjang, hal ini dapat menunjukkan adanya keterkaitan antara karakteristik morfologis penguin tersebut.

3. Dengan memperhatikan kedua grafik, kita dapat melihat adanya variasi yang signifikan dalam panjang paruh, massa tubuh, dan panjang sirip di antara penguin-penguin dalam dataset. Hal ini mengindikasikan adanya diversitas morfologis dalam populasi penguin yang diamati.

Analisis bivariate ini memberikan pemahaman tentang hubungan antara variabel-variabel yang diamati dalam dataset penguin, seperti panjang paruh, massa tubuh, dan panjang sirip. Analisis statistik yang lebih mendalam dan penelitian lebih lanjut dapat diperlukan untuk memahami keterkaitan dan signifikansi dari hubungan yang diamati dalam analisis ini.

# Soal 5. Multivariate Analisis

*   Dengan menggunakan seaborn, buatlah sebuah pairplot keseluruhan data feature pada dataset (df) dengan argumen kind='bar', seperti expected di bawah:
*   Jelaskan hasil analisa dari multivariate yang dihasilkan



```python
#code here
sns.pairplot(data=df)
plt.show()
```

Expected output:

Pair plot

![](https://drive.google.com/uc?id=1td8WAZLJEW5BPTq-wEJyZ4soRNNLXfqX)

Analisa:

Hasil analisa dari pairplot multivariate ini dapat memberikan beberapa insight tentang hubungan dan pola antara fitur-fitur dalam dataset. Berikut adalah beberapa hal yang dapat diperoleh dari pairplot:

Korelasi: Pairplot dapat menunjukkan korelasi antara pasangan fitur. Jika dua fitur memiliki pola linear atau hubungan yang terkait, scatter plot pada pairplot akan menunjukkan titik-titik data yang membentuk pola garis atau pola yang mengindikasikan hubungan positif atau negatif antara fitur-fitur tersebut.

Distribusi: Pairplot juga memperlihatkan distribusi dari masing-masing fitur pada diagonal utama. Histogram atau kernel density estimation (KDE) plot dapat memberikan gambaran tentang distribusi data dan apakah data cenderung terkonsentrasi di sekitar nilai tertentu atau memiliki variasi yang luas.

Outlier: Dalam pairplot, kita juga dapat melihat adanya outlier atau titik-titik data yang berada jauh dari pola umum. Outlier dapat memberikan insight tentang observasi yang tidak biasa atau potensial adanya kesalahan dalam data.

Variasi: Pairplot memperlihatkan variasi fitur-fitur dalam dataset. Dengan melihat pola hubungan antara fitur-fitur yang berbeda, kita dapat memperoleh pemahaman tentang variasi dalam dataset dan memperkirakan bagaimana fitur-fitur saling mempengaruhi.
